```python
df.fillna(df.mean(), inplace=True)
```

```python
fraud=df['Class'].value_counts()
label=["Not Fraud","Fraud"]
plt.figure(figsize=(10,8))
plt.pie(fraud,labels=label,autopct='%1.2f%%',startangle=90)
plt.show()
```

## The pie chart clearly shows that only 0.17% of the transactions in the dataset are fraudulent.
## This class imbalance highlights the need for special modeling strategies to detect rare but important fraud cases.
## Standard accuracy is not a reliable metric here — instead, precision, recall, and F1-score should be prioritized.

```python
# Separating features and target:

# 'x' contains all input features (excluding the target column 'Class')

# 'y' contains the target labels (0 = Not Fraud, 1 = Fraud)

x = df.drop('Class', axis=1)
y = df['Class']
```

```python
# Splitting the dataset into training and testing sets:

# 80% of the data will be used for training and 20% for testing

# 'random_state=7' ensures reproducibility (same split every time)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

## **This step is essential to evaluate the model’s performance on unseen (test) data.**

## **It helps us detect overfitting or underfitting.**

```python
len(x_train),len(x_test)
```

## **The dataset was successfully split into training and testing sets using an 80:20 ratio.**

## **Training set size: 227,845 samples**
## **New Section**
## **Testing set size: 56,962 samples**
## **This split allows the model to learn from a large portion of the data while keeping a sufficient amount for performance evaluation on unseen examples.**

```python
# df1=pd.DataFrame({
#     "Age" : [20,25,60,34,45,66],
#     "Salary" : [25000,20000,40000,50000,26000,70000]
# })

# df1
```

```python
# Feature Scaling using StandardScaler:
# This standardizes the features by removing the mean and scaling to unit variance.
# It is important for algorithms like Logistic Regression and KNN to perform well.
# 'fit_transform' is applied on training data, and 'transform' is applied on test data using the same scaling.

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

x_train_sc=sc.fit_transform(x_train) # Fit on training data and transform it
x_test_sc=sc.transform(x_test) # Only transform test data using same scaling
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import *
```

```python
# Training and evaluating Logistic Regression model:

# Step 1: Create a Logistic Regression model
model=LogisticRegression()

# Step 2: Train the model on the scaled training data
model.fit(x_train_sc,y_train)

# Step 3: Make predictions on the scaled test data
y_pred=model.predict(x_test_sc)

# Step 4: Calculate accuracy of the model
# This gives the percentage of correctly predicted instances out of all test instances
lr_accu=accuracy_score(y_pred,y_test)

print("Accuracy Score Is: ",lr_accu)
```

```python
# Confusion Matrix and classification report
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## Final Summary

- The dataset is highly imbalanced with only **0.17% fraud cases**.
- Logistic Regression achieved **overall high accuracy (≈ 100%)**.
- However, it detected only **56% of actual frauds** (recall = 0.56).
- Precision for fraud cases is **84%**, meaning most flagged frauds were correct.
- The model performs well, but **misses some fraud cases**, which is risky.

### Suggestion:
To improve fraud detection:
- Use **SMOTE** or other oversampling methods
